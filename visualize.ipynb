# Visualize the result

```python
import matplotlib.pyplot as plt
from collections import defaultdict

envs = ['maze2easy', 'maze2hard', 'ur5', 'snake7', 'kuka7', 'kuka13', 'kuka14']
envs_dimension = ['2D Easy', '2D Hard', 'UR5', 'Snake', 'Kuka7D', '13D', '14D']
metrics = ['success rate', 'collision check', 'running time', 'path cost', 'total time']
titles = ['Success Rate', 'Collision Checks', 'Running Time', 'Path Cost', 'Total Time']
metric_values = {}
method_names = {'GNN_ns': 'GNN',
                'GNN': 'GNN w/ Smoother',
                'GNN_pure_ns': 'GNN w/o Obstacle Encoding',
                'GNN_pure': 'GNN w/o Obstacle Encoding and w/ Smoother',
                'BIT*': 'BIT*',
                'NEXT': 'NEXT',
                'RRT*': 'RRT*', }
colors = ['rgb(246, 171, 171)', 'rgb(236, 86, 86)', 'rgb(192, 135, 191)', 'rgb(127, 15, 126)',
          'orange', '#2ca02c', 'rgb(100, 114, 246)', ]

with open("data/results/result__.txt", "r") as f:
    for line in f:
        result = line.strip()
        if len(set(result.split(' ')) & set(envs)) and 'Avg' in result:
            key = result.split()[:2]
            for metric in metrics:
                value = float(f.readline().split(': ')[1])
                metric_values[tuple([metric] + key)] = value
print(metric_values)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.clf()
plt.style.use('seaborn')

x = np.arange(len(method_names))  # the label locations
width = 0.1  # the width of the bars

fig, axes = plt.subplots(1, 4, figsize=(60, 15))
ax = axes[0]
for i, item in enumerate(method_names.items()):
    method, label = item
    rects1 = ax.plot(x, [metric_values['success rate', env, method] for env in envs], width, label=label)

#  - ((len(method_names)-1)*width)/2 + i * width
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)

fig.tight_layout()

handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=6)

plt.show()
```
